```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = 0.4
b = 0.5
c = 0.7
eta = 0.5
```

```python
r = np.array([a, b])
R = np.array([[1, c],[c, 1]])
sigma = 0.3
print(r)
print(R)
```

```python
def cost_f(sigma, r, w, R):
    return 0.5*sigma - np.dot(r.T, w)+0.5 * np.dot(np.dot(w.T, R), w)
def f(x, y):
    return 0.5*sigma**2 - a*x - b*y+ 0.5*(x**2+2*c*x*y+y**2)
```

```python
def gradient(N, eta, r, R, sigma):
    w_result = []
    cost_result = []
    w_0 = np.array([0, 0])
    w_result.append(w_0)
    cost_result.append(cost_f(sigma, r, w_0, R))
    for i in range(N):
        w_result.append(w_result[i] + eta * (r - np.dot(R, w_result[i])))
        cost_result.append(cost_f(sigma, r, w_result[i], R))
        
    return w_result, cost_result
```

```python
w, m = gradient(200, eta, r, R, sigma)
```

```python
t = np.array([1,1])
print(cost_f(sigma, r, t, R))
```

```python
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)

Z = f(X,Y)

fig = plt.figure(figsize = (5,5))
ax = plt.axes(projection='3d')

ax.scatter(2,3,4)
for x,y in w:
    ax.scatter(x,y,f(x,y))

ax.contour3D(X, Y, Z, 30, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
x1 = np.linspace(-1, 1, 150)
x2 = np.linspace(-1, 1, 150)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)
fig = plt.figure(figsize = (10,7))
plt.imshow(Z, extent = [-0.5,1.5,-1,1], origin = 'lower', cmap = 'jet', alpha = 1)
plt.title("Weight trajectory", fontsize=15)
plt.plot(np.array(w)[:,:1], np.array(w)[:,1:])
plt.plot(np.array(w)[:,:1], np.array(w)[:,1:], '*', label = "Cost function")

plt.xlabel('x1', fontsize=11)
plt.ylabel('x2', fontsize=11)
plt.colorbar()
plt.legend(loc = "upper right")
plt.show()
```
